# Demonstration of Covid-19 data

For the given covid-19 data:

    i.	Clean the data 
    ii.	performing geolocation plot by marking death, confirmed and recovered in marker. On clicking any specific location, it             should show the details in the map as shown in the image.


```python
import pandas as pd
confirmed=pd.read_csv('time_series_covid19_confirmed_global.csv')
death=pd.read_csv('time_series_covid19_deaths_global.csv')
recoverd=pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
new_df=confirmed.copy()
new_df.head()
```

```python
from folium import plugins
crime_cat = plugins.MarkerCluster().add_to(world_map)

for lat, lng, label, in zip(df3.Lat, df3.Long, df3.confirmed):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(crime_cat)


world_map
```
